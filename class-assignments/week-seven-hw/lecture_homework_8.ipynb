### 3d) Write a program that calculates the mean population for each city and year.

```python
mean_pop = {} 
# open an empty set, we will import values (dictionaries in this case) later
import csv 

with open('Dalziel2016_data.csv', 'r') as csv_file: 
    file = csv.DictReader(csv_file)
    for line in file:
        #will complete the following set of commands line-for-line
        loc = line['loc']
        # for each line, extract the value for variable 'loc'
        year = line['year']
        pop = float(line['pop']) 
        mean_pop[loc] = mean_pop.get(loc, {}) 
        #within the larger set, we will input dictionaries under the location names
        # as of right now, these dictionaries are empty, but the following code will add the key, value pair of 
        #year and number of weeks
        mean_pop[loc][year] = mean_pop[loc].get(year, [0,0])
        mean_pop[loc][year][0] = mean_pop[loc][year][0] + pop 
        #Total population
        mean_pop[loc][year][1] = mean_pop[loc][year][1] + 1 
        #Number of weeks (incrementing by 1)

#the following for loop will get the year (key) per location/city (dictionary)
# This is where we actually calculate the average population according to these variables
        
for city in mean_pop.keys():
    for year in mean_pop[city].keys():
        mean_pop[city][year][0] = mean_pop[city][year][0]/mean_pop[city][year][1] 
        #Total population / # of weeks    

#the following for loop will iput the city name in mean_pop to generate the average population per 
#location and year        
        
cities = ['LOS ANGELES', 'CLEVELAND', 'SAN FRANCISCO', 'RICHMOND', 'TOLEDO', 'DULUTH', 'BALTIMORE', 'NEWARK', 'KANSAS CITY', 'ROCHESTER', 'CINCINNATI', 'CHICAGO', 'SPRINGFIELD', 'BOSTON', 'NASHVILLE', 'MINNEAPOLIS', 'MILWAUKEE', 'NEW ORLEANS', 'WORCESTER', 'HARTFORD', 'NEW HAVEN', 'GRAND RAPIDS', 'BRIDGEPORT', 'FALL RIVER', 'SPOKANE', 'BUFFALO', 'NEW YORK', 'INDIANAPOLIS', 'COLUMBUS', 'ST LOUIS', 'WASHINGTON', 'SEATTLE', 'TRENTON', 'PITTSBURGH', 'READING.US', 'PROVIDENCE', 'PHILADELPHIA', 'DETROIT', 'DENVER', 'SALT LAKE CITY']

for city in cities:
    years = list(mean_pop[city].keys())
    years.sort()
    print("City: ", city)
    for year in years:
        print(year, round(mean_pop[city][year][0]))
    print('\n')
```

### 4) Singh et al. (2015) show that, when infected with a parasite, the four genetic lines of D. melanogaster respond by increasing the production of recombinant offspring (arguably, trying to produce new recombinants able to escape the parasite). They show that the same outcome is not achieved by artificially wounding the flies. The data needed to replicate the main claim (figure 2 of the original article) is contained in the file Singh2015_data.csv in your data directory. Open the file, and compute the mean RecombinantFraction for each Drosophila Line, and InfectionStatus (W for wounded and I for infected). 

### Print the results in the following form:
1
Line 45 Average Recombination Rate:
W : 0.187
I : 0.191

```python
import csv 
my_data = {}
with open('Singh2015_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        my_line = row['Line']
        my_status = row['InfectionStatus']
        my_recomb = float(row['RecombinantFraction'])
        if my_line not in my_data:
            my_data[my_line] = {'W': [], 'I': []}
        my_data[my_line][my_status].append(my_recomb)

for line in my_data:
    print('Line', line, 'Average Recombination Rate:')
    my_subset = my_data[line]
    for status in ['W', 'I']:
        print(status, ':', end = '') 
        my_mean = sum(my_subset[status])
        my_num_elements = len(my_subset[status])
        my_mean = my_mean / my_num_elements
        print(' ', round(my_mean, 3))
    print('')
```
