```python
import numpy as np
import pandas as np
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples = 300, centers = 3, cluster_std = 0.60, random_state = 0)
plt.scatter(X[:,0], X[:,1], c=y)
# c=y for color, otherwise just blue
```

```python
X, y = make_blobs(n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 0)
# changed # of clusters
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
wcss = []
k_min = 1
k_max = 10
for kval in range(k_min, k_max+1):
    kmeans = KMeans(n_clusters = kval, init = "k-means++", max_iter = 300, 
                     n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
print(wcss)
```

```python
plt.plot(range(k_min, k_max+1), wcss)
plt.title("Elbow Method")
plt.xlabel('Number of clusters (k)')
plt.ylabel("WCSS")
```

```python
kmeans = KMeans(n_clusters = 4, init = "k-means++", max_iter=300, n_init = 10, rando_state = 0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c= pred_y)
# overlay:
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_center_[:,1],
           s = 300, c= "red")
plt.show()
```
