```python
ave_inflammation = np.mean(data, axis=0)
ave_plot = plt.plot(ave_inflammation)
plt.xlabel('time (days)')
plt.ylabel('inflammation score')
plt.title('patient data')
plt.show()
```

```python
max_plot = plt.plot(np.max(data,axis=0))
plt.show()
#max gives you sharper peaks
```

```python
min_plot = plt.plot(np.min(data, axis = 0))
plt.show
```

```python
#Group plots function into subplots
#created a figure (object that holds mult plots)
fig = plt.figure(figsize=(10.0 ,3.0))

#1 row, 3 col, index 1, 2, 3
#3 placeholder variables --> tell it where you wanna store stuff
sbplt1 = fig.add_subplot(1, 3, 1)
sbplt2 = fig.add_subplot(1, 3, 2)
sbplt3 = fig.add_subplot(1, 3, 3)

sbplt1.set_ylabel('average')
sbplt1.plot(np.mean(data, axis=0), '--')
#'--' makes the line dashed, you can do this with bullets and stars too

sbplt2.set_ylabel('max')
sbplt2.set_xlabel('time(days)')
sbplt2.plot(np.max(data, axis=0))

sbplt3.set_ylabel('min')
sbplt3.plot(np.min(data, axis=0))

fig.tight_layout()
#adjustment function --> uniform format between graphs

plt.show()
```

```python
#get all even numbers from 1 to 100
evens = []
for ii in range(1,101):
    if (ii % 2 == 0):
        evens.append(ii)
print(evens)

#get all odd numbers from 1 to 100
odds = []
for ii in range(1,101):
    if (ii % 2 == 1):
        odds.append(ii)
print(odds)
```

```python
#find squares from first 10 even numbers in the sequence
squares_nums = [i * i for i in range(20) if i % 2 == 0]
print(squares_nums)
```
